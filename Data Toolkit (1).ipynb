1.What is NumPy, and why is it widely used in Python?

Ans:NumPy is a core Python library for numerical computing. It provides fast, efficient operations on large, multi-dimensional arrays and matrices. Widely used for its speed, vectorized operations, and support for broadcasting, NumPy is the foundation for many data science and machine learning libraries like Pandas, Scikit-learn, and TensorFlow.

2.How does broadcasting work in NumPy?

Ans:Broadcasting in NumPy lets you perform operations on arrays of different shapes by automatically expanding the smaller array to match the larger one. It follows simple rules to align dimensions and avoids the need for manual reshaping, making computations faster and cleaner.

3.What is a Pandas DataFrame?

Ans:A Pandas DataFrame is a 2-dimensional, labeled data structure in Python, similar to a table in a database or an Excel spreadsheet. It consists of rows and columns, where each column can hold different data types (e.g., int, float, string). It’s widely used for data analysis and manipulation due to its easy handling of missing data, filtering, grouping, and powerful built-in functions.

4.Explain the use of the groupby() method in Pandas.

Ans:The groupby() method in Pandas is used to split data into groups based on one or more columns, allowing you to perform operations like aggregation, transformation, or filtering on each group separately. It's especially useful for summarizing data.

Example:

df.groupby('Category')['Sales'].sum()

This groups the DataFrame by "Category" and calculates the total "Sales" for each group.
In short, groupby() is powerful for analyzing patterns and statistics across different segments of data.

6.What are the differences between NumPy arrays and Python lists?

Ans:NumPy arrays differ from Python lists in several important ways. NumPy arrays are faster and more memory-efficient, as they store elements of the same data type in contiguous memory blocks. In contrast, Python lists can hold mixed data types but are slower and use more memory. NumPy also supports advanced features like mathematical operations, broadcasting, and slicing, which are not available with regular lists. Overall, NumPy arrays are better suited for numerical and scientific computing, while Python lists are more general-purpose.

7.What is a heatmap, and when should it be used?

Ans:A heatmap is a data visualization tool that uses color to represent values in a matrix or table. Each cell's color intensity reflects the magnitude of the value, making it easy to spot patterns, correlations, or anomalies in the data. Heatmaps are commonly used for visualizing correlation matrices, confusion matrices, or large datasets where understanding the relationship between variables is important. They are especially useful in exploratory data analysis and feature selection.

8.What does the term “vectorized operation” mean in NumPy?

Ans:In NumPy, a vectorized operation refers to performing operations on entire arrays without using explicit loops. Instead of processing each element one by one, NumPy applies the operation to all elements at once using optimized C-based code under the hood. This makes the code faster, cleaner, and more efficient. For example, adding two arrays element-wise can be done directly with a + b , without looping through each element.

9.A How does Matplotlib differ from Plotly?

Ans:Matplotlib and Plotly are both popular Python libraries for data visualization, but they serve different purposes. Matplotlib is primarily used for creating static, publication-quality plots and offers extensive customization options. It's well-suited for scientific research and traditional graphs but lacks built-in interactivity. On the other hand, Plotly is designed for creating interactive, web-based visualizations with features like zooming, hovering, and tooltips. It is ideal for dashboards and modern data exploration tools. In summary, Matplotlib is best for static visuals, while Plotly excels at creating interactive and dynamic plots.

10.What is the significance of hierarchical indexing in Pandas?

Ans:Hierarchical indexing in Pandas, also known as MultiIndex, allows you to have multiple levels of indexes on a DataFrame or Series. This is especially useful for working with high-dimensional data in a two-dimensional structure. It enables more flexible data selection, aggregation, and reshaping.

For example, you can group data by multiple keys (like region and year) and easily perform operations across different levels. It also supports advanced reshaping methods like stack() and unstack(). In short, hierarchical indexing helps organize and analyze complex datasets more efficiently.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy’s array slicing is faster because it returns views of the original data rather than creating new copies like Python lists do. NumPy arrays use contiguous memory blocks and are implemented in C, which allows faster access and slicing. Also, NumPy uses fixed data types, enabling optimized operations, unlike Python lists which are more flexible but slower due to dynamic typing and object references.

```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Marks': [85, 90, 78, 92]}

df = pd.DataFrame(data)

# Find the mean of the 'Marks' column
mean_marks = df['Marks'].mean()

print("Mean Marks:", mean_marks)
```

```python
#3.Create a scatter plot using Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Height': [150, 160, 170, 180, 190],
    'Weight': [50, 60, 65, 80, 90]
}

df = pd.DataFrame(data)

# Create scatter plot
plt.scatter(df['Height'], df['Weight'], color='green', marker='x')

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight Scatter Plot')

# Show the plot
plt.show()
```

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'Math': [90, 85, 78, 92],
    'Physics': [88, 79, 85, 91],
    'Chemistry': [84, 82, 80, 88],
    'Biology': [86, 80, 83, 89]
})

# Correlation matrix
corr = df.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

```

```python
 #5.Generate a bar plot using Plotly
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [15000, 18000, 12000, 20000],
    'op': [3000, 4000, 2500, 5000]
}

df = pd.DataFrame(data)

# Convert to long format
df_long = df.melt(id_vars='Region', value_vars=['Sales', 'op'],
                  var_name='Metric', value_name='Amount')

# Create grouped bar plot
fig = px.bar(df_long, x='Region', y='Amount', color='Metric',
             barmode='group', title='Sales and OP by Region')

# Show plot
fig.show()
```

```python
#8.A Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Sample data
years = [2018, 2019, 2020, 2021, 2022]
sales_A = [200, 250, 300, 350, 400]
sales_B = [180, 220, 260, 310, 370]

# Create line plot
plt.plot(years, sales_A, label='Product A', color='blue', marker='o')
plt.plot(years, sales_B, label='Product B', color='green', marker='s')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales of Product A and B')

# Add legend
plt.legend()

# Show plot
plt.show()

```

```python
#10.Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'Marks': [55, 65, 70, 70, 72, 75, 78, 80, 82, 85, 88, 90, 91, 92, 95]}
df = pd.DataFrame(data)

# Create histogram
sns.histplot(df['Marks'], bins=8, kde=True, color='skyblue')

# Add labels and title
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')

# Show plot
plt.show()
```
